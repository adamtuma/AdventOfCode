```python
### PART 1 ###

txt_file = open('day2_input.txt', 'r')
data = [line.strip().split(" ") for line in txt_file]
txt_file.close()

df = pd.DataFrame(data, columns = ["command", "value"]).apply(pd.to_numeric, errors = "ignore")
df1 = df.groupby("command").sum()

answer1 = int((df1.loc["down"]-df1.loc["up"])*df1.loc["forward"])
print(f"Answer for part one is: {answer1}")   

### PART 2 ###
aim = 0
df["aim"] = 0
for i in range(0,len(df)):
    if df.loc[i,"command"] == "up":
        aim = aim - df.loc[i,"value"]
    elif df.loc[i,"command"] == "down":
        aim = aim + df.loc[i,"value"]
    else:
        pass
    df.loc[i,"aim"] = aim

df["vertical"] = 0
for i in range(0,len(df)):
    if df.loc[i,"command"] == "forward":
        df.loc[i,"vertical"] = df.loc[i,"value"]*df.loc[i,"aim"]
    else:
        pass
        
answer2 = int(df1.loc["forward"]*df["vertical"].sum())
print(f"Answer for part two is: {answer2}")   
```

```python
### PART 1 ###
txt_file = open('day8_input.txt', 'r')
data = [line.strip() for line in txt_file]
txt_file.close()

data1 = [line.split(" | ") for line in data]
data1 = [line[1].split(" ") for line in data1]

answer1 = 0
values = [2,3,4,7]
for i in data1:
    for k in i:
        if len(k) in values:
            answer1 += 1
        else:
            pass
print(f"Answer for part one is: {answer1}")

### PART 2 ###
data0 = [line.split(" | ") for line in data]
data0 = [line[0].split(" ") for line in data0]

[i.sort(key=len) for i in data0]
data0

def create_sets(sorted_list):
    set_1 = set(sorted_list[0])         #assigning basic values based on length
    set_7 = set(sorted_list[1])
    set_4 = set(sorted_list[2])
    set_8 = set(sorted_list[-1])
    for i in sorted_list[6:9]:          #assigning values for 5-segment numbers
        if set_4.issubset(set(i)):      #using the fact that 9 contains all segments of 4 
            set_9 = set(i)
        elif set_1.issubset(set(i)):    #0 contains segments of 1
            set_0 = set(i)
        else:                           #only 6 is left
            set_6 = set(i)
    for j in sorted_list[3:6]:
        if set_1.issubset(set(j)):      #3 contains segments of 1
            set_3 = set(j)
        elif set(j).issubset(set_6):    #5 contains all segments of 6
            set_5 = set(j)
        else:                           #only 2 is left
            set_2 = set(j)
    sets = [set_0,set_1,set_2,set_3,set_4,set_5,set_6,set_7,set_8,set_9]
    return sets

def match_values(input_list, output_list):
    outputs = []
    for n,i in enumerate(input_list):
        sets = create_sets(i)
        output = ""
        for j in output_list[n]:
            k = 0
            while set(j) != sets[k]:    #the position when there is a hit is also the number 
                k += 1
            output += str(k)
        outputs.append(int(output))
    return outputs

output_numbers = match_values(data0,data1)
print(f"Answer for part two is: {sum(output_numbers)}")

```

```python
### PART 1 ###
txt_file = open('day10_input.txt', 'r')
data = [line.strip() for line in txt_file]
txt_file.close()

def find_illegal(string):
    openings = ["(","[","{","<"]
    closings = [")","]","}",">"]
    legal = "YES"
    n = 0
    ordered_openings = []
    possible_closing = ""
    while legal == "YES":
        try:
            if string[n] in openings:
                ordered_openings.append(string[n])
            else:
                if string[n] == possible_closing:
                    del ordered_openings[-1]
                else:
                    legal = "NO"
                    illegal_character = string[n]
            n += 1
            possible_closing = closings[openings.index(ordered_openings[-1])]
        except:
            illegal_character = ""
            break
    return illegal_character

syntax_error_score = 0    
for i in data:
    illegal_char = find_illegal(i)
    if illegal_char == ")":
        syntax_error_score += 3
    elif illegal_char == "]":
        syntax_error_score += 57
    elif illegal_char == "}":
        syntax_error_score += 1197
    elif illegal_char == ">":
        syntax_error_score += 25137
    else:
        pass

print(f"Answer for part one is: {syntax_error_score}")

### PART TWO ###

def get_score(string):              #adjusting the function to suit part 2
    openings = ["(","[","{","<"]
    closings = [")","]","}",">"]
    scores = [1,2,3,4]
    total_score = 0
    legal = "YES"
    n = 0
    ordered_openings = []
    possible_closing = ""
    while legal == "YES":
        try:
            if string[n] in openings:
                ordered_openings.append(string[n])
            else:
                if string[n] == possible_closing:
                    del ordered_openings[-1]
                else:
                    legal = "NO"
                    illegal_character = string[n]
            n += 1
            possible_closing = closings[openings.index(ordered_openings[-1])]
        except:
            illegal_character = ""
            ordered_closings = []
            while len(ordered_openings) >= 1:
                ordered_closings.append(closings[openings.index(ordered_openings[-1])])
                del ordered_openings[-1]
            for i in ordered_closings:
                total_score = total_score*5 + scores[closings.index(i)]
            break
    return total_score

scores = []
for i in data:
    score = get_score(i)
    if score > 0:
        scores.append(score)
    else:
        pass

print(f"Answer for part two is: {int(np.median(np.array(scores)))}")
```
